# Reading the data and storing them into dataframes

```python
def read_data(fpath):
    xlsx = pd.ExcelFile(fpath) # reading the excel file
    data_frame = [0 for names in xlsx.sheet_names] # making a list so that suits number of dataframes
    # reading the excel file sheet by sheet and storing each sheet in a separate pandas dataframe 
    for i in range(0,len(xlsx.sheet_names)):
        data_frame[i] = xlsx.parse(xlsx.sheet_names[i], skipfooter=1) #skipfooter to drop the last row
        data_frame[i] = data_frame[i].iloc[:,:-7] # to drop the last column and saving under 18 data
        data_frame[i].sheet_name = xlsx.sheet_names[i] # adding sheet name attribute to the pandas dataframe object
    return data_frame

fpath = 'Players_Data.xlsx'
data_frame = read_data(fpath)
```

```python
def get_club_data(data_frame):
    
    n = int(len(data_frame)/2)
    df_club = []
    for i in range(0,len(data_frame)):
        if i%2==0:
            df_club.append(data_frame[i])
    return df_club
```

```python
def get_player_data(data_frame):
    n = int(len(data_frame)/2)
    df_player = []
    for i in range(0,len(data_frame)):
        if i%2==1:
            df_player.append(data_frame[i])
    return df_player
df_play = get_player_data(data_frame)
```

```python
def get_players_count(df_play):
    players_count = []
    for i in range(0,len(df_play)):
        players_count.append(df_play[i].groupby(df_play[i]['Constituent Body']).sum())
        players_count[i].sheet_name = df_play[i].sheet_name
    return players_count
```

```python
def get_total_count(groupped_df):
    for i in range(0,len(groupped_df)):
        groupped_df[i]['Total'] = groupped_df[i][list(groupped_df[i].columns)].sum(axis=1)
    return groupped_df
```

```python
def get_players_data_stats(df_play):
    players_stat = []
    for i in range(0,len(df_play)):
        players_stat.append(df_play[i].describe())
        players_stat[i].sheet_name = df_play[i].sheet_name
    return players_stat
```

```python
def get_age_data(age, df_play):
    res = []
    doc_str = []
    col_str_m = 'U' + str(age) + 'M'
    col_str_f = 'U' + str(age) + 'F'
    for i in range(1,len(df_play)):
        res.append(df_play[i][['Constituent Body', str(col_str_m), str(col_str_f)]])
        res[i-1].sheet_name = df_play[i].sheet_name
        doc_str.append(str(age) + ' years old players\' data from the ' + df_play[i].sheet_name + ' sheet')
    return res
```

```python
def data_flow(age, df_play):
    org_age = age
    ages = [j for j in df_play[0].columns if j.startswith('U')]
    res = []
    doc_str = []
    col_str_m = 'U' + str(org_age) + 'M'
    col_str_f = 'U' + str(org_age) + 'F'
    for i in range(1,len(df_play)):
        res.append(df_play[i][['Constituent Body', str(col_str_m), str(col_str_f)]])
        res[i-1].sheet_name = df_play[i].sheet_name
        doc_str.append(str(age) + ' years old players\' data from the ' + df_play[i].sheet_name + ' sheet')
        age +=1
        if age>17:
            return res
        col_str_m = 'U' + str(age) + 'M'
        col_str_f = 'U' + str(age) + 'F'
    return res
```

```python
def get_gross_figures(df):
    gross_number_by_age = df.filter(regex='U').sum(axis=0)
    gross_number_by_sex = (gross_number_by_age.filter(regex='F').sum(), gross_number_by_age.filter(regex='M').sum())
    gross_number_total = gross_number_by_sex[0] + gross_number_by_sex[1]
    return gross_number_by_age, gross_number_by_sex, gross_number_total
```

```python
def get_the_whole_gross(df_play):
    annual_gross = [[0 for i in range(0,3)] for j in range(0,len(df_play))]
    for i in range(1, len(df_play)):
        _, gross_number_by_sex, gross_number_total = get_gross_figures(df_play[i])
        annual_gross[i][0] = gross_number_by_sex[0]
        annual_gross[i][1] = gross_number_by_sex[1]
        annual_gross[i][2] = gross_number_total
    del annual_gross[0]
    return pd.DataFrame(data = annual_gross, index = ['2016', '2017', '2018', '2019', '2019-Nov', '2021-Jan', '2023-May'], columns=['Female', 'Male', 'Total'])
get_the_whole_gross(df_play)
```

```python
def gross_bar_plot(df_play):
    get_the_whole_gross(df_play).iloc[:, :].plot.bar(subplots=True)
gross_bar_plot(df_play)
```
